#Task 1: Graph a function of one variable Plot the graph of sin(x) for x in the interva [0,10].

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Generate x values from 0 to 10 with a certain number of points
x = np.linspace(0, 10, 100)  # You can adjust the number of points for a smoother or more detailed plot

# Compute the corresponding y values (sin(x))
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.plot(x, y, label='sin(x)')
plt.title('Plot of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Task 2: Graph a function of two variables Plot the graph of cos( x2 + y2)^(1/2) for (x, y)[-5, 5]  [-5, 5].

```python
# Define the range for x and y
x = np.linspace(-5, 5, 100)  # 100 points in the range [-5, 5]
y = np.linspace(-5, 5, 100)  # 100 points in the range [-5, 5]

# Create a grid of (x, y) values
X, Y = np.meshgrid(x, y)

# Calculate the function values for each (x, y) pair
Z = np.cos(np.sqrt(X**2 + Y**2))

# Create a contour plot
plt.contourf(X, Y, Z, levels=100, cmap='viridis')
plt.colorbar()

# Add labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of cos(sqrt(x^2 + y^2))')

# Display the plot
plt.show()
```
