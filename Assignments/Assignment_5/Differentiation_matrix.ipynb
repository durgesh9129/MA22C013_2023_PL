```python
import numpy as np
```

```python
n = 1000;
dif_fun = np.zeros((n-1,n))
for i in range(n-1):
  for j in range(n):
    if (i ==j):
      dif_fun[i,j] = 1
    elif (j == i-1):
      dif_fun[i,j] = -1
print(dif_fun)
```

```python
dif_fun.shape
```

```python
n = 1000
x = np.linspace(0, 2*3.14, 1000)
y = np.sin(x)
y1 = np.matmul(dif_fun,y)
y1 = y1/(2*3.14/1000)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(x[1:],y1,label ='cosx')
plt.plot(x,y,label ='sinx')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2*3.14, 1000)
y = x**2 + 2*(np.sin(10*x))
y1 = np.matmul(dif_fun,y)
y1 = y1/(2*3.14/1000)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(x[1:],y1,label ='derivative of F(x)')
plt.plot(x,y,label ='F(x)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
from numpy import diag
import matplotlib.pyplot as plt
```

```python
def diff_funs(n):
  diff1 = np.diag(np.ones((n-2)),-1)
  diff2 = np.diag(np.ones((n-1)))
  diff_mat = diff2 - diff1
  return diff_mat
```

```python
diff_mat = diff_funs(n)
```

```python
x_new = x[:-1]
y_new = x_new**2 + 2*(np.sin(10*x_new))
y_new.shape
```

```python
y_new = x_new**2 + 2*(np.sin(10*x_new))
y1_new = np.matmul(diff_mat,y_new)
y1_new = y1_new/(2*3.14/1000)
```

```python
plt.plot(x_new,y1_new,label ='derivative of F(x)')
plt.plot(x,y,label = 'F(x)')
plt.legend(loc='best')
```
