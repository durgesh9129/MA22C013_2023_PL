```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 1000001)
```

```python
f=np.piecewise(x, [x < 0, x >= 0], [lambda x: x*np.sin(x*2), lambda x: np.cos(x)])
f
```

```python
%timeit np.piecewise(x, [x < 0, x >= 0], [lambda x: x*np.sin(x*2), lambda x: np.cos(x)])
```

```python
plt.plot(x,f)
plt.grid(True)
```
